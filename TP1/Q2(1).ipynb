### 1)
Lets define $$ \hat{p} = \frac{1}{m}\sum_{i}{Z_i} $$

So taking the exponent and raising to the power of m we get:
$$ P(\hat{p} \geq p + \epsilon) = P(e^{\lambda\hat{p}}\geq e^{\lambda(p+\epsilon)}) = P(e^{\lambda\sum_{i}{Z_i}}\geq e^{m\lambda(p+\epsilon)}) $$
Further applying the inequality of Markow ad using the hint given we get:
$$ P(\hat{p} \geq p + \epsilon) \leq \frac{\mathbb{E}[e^{\lambda\sum_{i}{Z_i}}]}{e^{m\lambda(p+\epsilon)}} = \frac{\mathbb{E}[e^{\lambda Z}]^{m}}{e^{m\lambda(p+\epsilon)}} $$
Using the expeectation of a Bernoulli variable: $ \mathbb{E} [e^{\lambda Z}] = pe^\lambda + (1-p)$ We get:
$$ P(\hat{p} \geq p + \epsilon) \leq (\frac{pe^\lambda + (1-p)}{e^{\lambda(p+\epsilon)}})^m $$

### 2)
Using the hint, we first take the log of the right hand side we get: 

$$
\log{f(p,\epsilon,\lambda)}\ =\ \log{((\frac{pe^\lambda + (1-p)}{e^{\lambda(p+\epsilon)}})^m)}\ =\ m\log{(\frac{pe^\lambda + (1-p)}{e^{\lambda(p+\epsilon)}})}\\
=>m(\log{(pe^\lambda + (1-p))}-\log{(e^{\lambda(p+\epsilon)})})\ =\ m\log{(pe^\lambda + (1-p))} - m(\lambda(p+\epsilon))
$$


To find the minimum we take the point where the derivative of the expression above with respect to $\lambda$ is equal to 0:


$$
\begin{align}
& \frac{d}{d\lambda}f(p,\epsilon,\lambda^*)\ &&=\ 0\\
=>\ & m(\frac{pe^{\lambda^*}}{(pe^{\lambda^*} + (1-p))})-m(p+\epsilon)\ &&=\ 0\\
=\ & \frac{pe^{\lambda^*}}{(pe^{\lambda^*} + (1-p))} &&= (p+\epsilon)\\
=\ & pe^{\lambda^*} &&= (p+\epsilon)(pe^{\lambda^*} + (1-p))\\
=\ & e^{\lambda^*} p(1-(p+\epsilon)) &&= (p+\epsilon) (1-p)\\
=\ & e^{\lambda^*} &&= \frac{(p+\epsilon) (1-p)}{p(1-(p+\epsilon))}
\end{align}
$$

Plugging this in the inequality gives us the wanted:

$$
\begin{align}
P\left(\hat{p} \geq p + \epsilon\right) \leq\ &\left(\frac{pe^\lambda + (1-p)}{e^{\lambda(p+\epsilon)}}\right)^m \\
\leq\ & \left(\frac{(p\frac{(p+\epsilon)) (1-p)}{p(1-(p+\epsilon))} + (1-p)}{e^{\log{\left(\frac{(p+\epsilon) (1-p)}{p(1-(p+\epsilon))}\right)}(p+\epsilon)}}\right)^m\\
\leq\ & \left(\frac{\frac{(p+\epsilon) (1-p)}{1-(p+\epsilon)} + \frac{(1-p)1-(p+\epsilon)}{1-(p+\epsilon)}}{e^{\log{\left(\frac{(p+\epsilon) (1-p)}{p(1-(p+\epsilon))}\right)}(p+\epsilon)}}\right)^m\\
\leq\ & \left(\frac{\frac{(1-p)}{1-(p+\epsilon)} }{e^{\log{\left(\frac{(p+\epsilon) (1-p)}{p(1-(p+\epsilon))}\right)}(p+\epsilon)}}\right)^m\\
\leq\ & \left(\frac{e^{\log{\left(\frac{(1-p)}{1-(p+\epsilon)}\right)}} }{e^{\log{\left(\frac{(p+\epsilon) (1-p)}{p(1-(p+\epsilon))}\right)}(p+\epsilon)}}\right)^m\\
\leq\ & \left(e^{\left(\log{\left(\frac{(1-p)}{1-(p+\epsilon)}\right)}- \log{\left(\frac{(1-p)(p+\epsilon)}{p(1-(p+\epsilon))}\right)}(p+\epsilon)\right)m}\right) \\
\leq\ & \left(e^{\left(\log{\left(\frac{(1-p)}{1-(p+\epsilon)}\right)}- (p+\epsilon)\log{\left(\frac{1-p}{1-(p+\epsilon)}\right)}-(p+\epsilon)\log{\left(\frac{p+\epsilon}{p}\right)}\right)m}\right) \\
\leq\ & \left(e^{\left(\log{\left(\frac{(1-p)}{1-(p+\epsilon)}\right)} \left(1 - (p+\epsilon)\right)-(p+\epsilon)\log{\left(\frac{p+\epsilon}{p}\right)}\right)m}\right) \\
\leq\ & \left(e^{\left(- \log{\left(\frac{1-(p+\epsilon)}{(1-p)}\right)} \left(1 - (p+\epsilon)\right) - (p+\epsilon)\log{\left(\frac{p+\epsilon}{p}\right)}\right)m}\right) \\
\leq\ & \left(e^{-\left(\log{\left(\frac{1-(p+\epsilon)}{(1-p)}\right)} \left(1 - (p+\epsilon)\right) + (p+\epsilon)\log{\left(\frac{p+\epsilon}{p}\right)}\right)m}\right) \ =\ e^{-mf(p,\epsilon)} \\
\end{align}
$$

### 3)

We have that:

$$
f(p,\epsilon)\ = \ \log{\left(\frac{1-(p+\epsilon)}{(1-p)}\right)} \left(1 - (p+\epsilon)\right) + (p+\epsilon)\log{\left(\frac{p+\epsilon}{p}\right)}
$$

- $$
f(p,0)\ =\ \log{\left(\frac{1-(p+ 0 )}{(1-p)}\right)} \left(1 - (p+ 0)\right) + (p+0)\log{\left(\frac{p+0}{p}\right)}\\ f(p,0)\ =\ 0 
$$

- $$
\frac{df(p,\epsilon)}{d\epsilon}\big|_{\epsilon = 0} = \log{\left(\frac{p+\epsilon}{p}\right)}+\frac{p}{p+\epsilon}\frac{p+\epsilon}{p}-\log{\left(\frac{1-(p+\epsilon)}{1-p}\right)} -\frac{1}{1-p}\frac{1-p}{(1-(p+\epsilon))}(1-(p+\epsilon))\ \big|_{\epsilon = 0}\\
= \log{\left(\frac{p+\epsilon}{p}\right)} + 1 - \log{\left(\frac{1-(p+\epsilon)}{1-p}\right)} - 1\ \big|_{\epsilon = 0} \\
= 0
$$
- For this part we equal the third derivaive to zero in order to find the maxima 
$$\begin{align}
\frac{d^2f(p,\epsilon)}{d\epsilon^2}\ &=\ \frac{1}{p}\frac{p}{p+\epsilon}+\frac{1}{1-p}\frac{1-p}{1-(p+\epsilon)}\\
&=\ \frac{1}{p+\epsilon}\frac{1}{1-(p+\epsilon)} \\
\frac{d^3f(p,\epsilon)}{d\epsilon^3}\ &=\ \frac{2(p+\epsilon)+1}{(p+\epsilon-1)^2(x+p)^2}\ = 0 \\
\ &=\ (p+\epsilon) = \frac{1}{2}
\end{align}\\
=>\frac{d^2f(p,\epsilon)}{d\epsilon^2} \big|_{(p+\epsilon)= \frac{1}{2}}\ = 4
$$

### 4)
Using what we had from question 3 we get the taylor expression:

$$
\begin{align}
f(p, \epsilon)\ &=\ f(p,0) +\epsilon f'(p,0) + \frac{\epsilon ^2}{2}f''(p,\tilde{\epsilon})\\
& =\ 0 + 0 + \frac{\epsilon ^2}{2}f''(p,\tilde{\epsilon})
\end{align}
$$

as we know that  $ \frac{d^2f(p,\epsilon)}{d\epsilon^2} \geq 4 $ for the right hand side this results in: 

$$
f(p, \epsilon)\ \geq \frac{\epsilon ^2}{2} \cdot 4\ =>\ f(p, \epsilon)\ \geq \epsilon ^2 2
$$

With this inequality we can conclude the proof of (5) as the minimum value of $f(p,\epsilon)$ is $\epsilon ^2 2$, plugging it into the equation from question 2 we get the desired:
$$
P\left(\frac{1}{m}\sum_{i}Z_i \geq p+ \epsilon \right) \leq e^{-2m\epsilon ^2}
$$

```python
import numpy as np
import pandas as pd
from scipy.stats import bernoulli
from scipy import stats
import matplotlib.pyplot as plt
p_lis = [0.2, 0.5, 0.8]
m_theory = 18444

def ber(p, m):
    elements = bernoulli.rvs(p, size = m)
    return elements

err = []
count = 0
count_m_the = []
for x, p in enumerate(p_lis):
    for i in range(1, 100):
        elements = ber(p,m_theory)
        p_estim = elements.sum()/len(elements)
        err = (p-p_estim)/p
        if err < 0.01:
            count = count+1
    print(f"p =",p, ",probability of error < 1%:",count/100)
    count_m_the.append(count/100)
    count = 0
```

We can see that the percentage which the error is less than 1% is smaller for p small, non the less the bounds are  accurate. 

```python
k = np.linspace(1, 10, 10)
m_range = m_theory*k
err = []
count = 0
for p in p_lis:
    for m in m_range:
        for i in range(1, 100):
            elements = ber(p,int(m))
            p_estim = elements.sum()/len(elements)
            err = (p-p_estim)/p
            if err < 0.01:
                count = count+1
        print(f"p =",p, ",k =", np.where(m_range == m)[0], ",probability of error < 1%:",count/100)
        elements = []
        count = 0
```

After comparing results with different coefficiants, we should keep an eye on our lowest p = 0.2. We observe that after k = 6 meaning 6 times the theoretically calculated value of m, the error percentage for each p results higher than 95%.
